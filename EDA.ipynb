```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris_data = sns.load_dataset('iris')
```

```python
iris_data
```

```python
iris_data['species'].value_counts()
```

```python
sns.histplot(iris_data, x = 'petal_length', hue = 'species')
```

```python
sns.histplot(iris_data, x = 'sepal_length', hue = 'species')
```

```python
sns.kdeplot(iris_data, x = 'petal_length', hue = 'species')
```

# hume hrr chiz yaani hrr feature k liye plot bnana pdta h to more explore the data|
- or kai baar single feature se proper clarity ni dikhti to 2,3 feature ikathe plot krke dekhne pdte h

```python
sns.kdeplot(iris_data, x = 'petal_length',y = 'petal_width', hue = 'species')
```

```python
sns.histplot(iris_data, x = 'petal_length',y = 'petal_width', hue = 'species')
```

```python
sns.scatterplot(iris_data, x = 'petal_length',y = 'petal_width', hue = 'species')
```

```python
sns.jointplot(iris_data, x = 'petal_length',y = 'petal_width', hue = 'species')
```

```python
sns.pairplot(iris_data, hue = 'species')
```
